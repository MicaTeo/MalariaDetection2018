# HELPER FUNCTIONS

```python

def train_evaluate_knn(train_X, train_y, test_X, test_y, n_test, n_datas, ks, expName, save_path, plot_ind=False, save_individual=False):
    roc_aucs = np.zeros((len(n_datas), len(ks)))
    prc_aucs = np.zeros((len(n_datas), len(ks)))
    # Training
    for j, k in enumerate(ks):
        for i, n_data in enumerate(n_datas):
            clf = neighbors.KNeighborsClassifier(n_neighbors=k, weights='distance')
            clf.fit(train_X[:n_data].reshape((n_data,-1)), train_y[:n_data].reshape((n_data,-1)))
            pred_y = clf.predict_proba(test_X[:n_test].reshape((n_test,-1)))[:,1]

            # Plotting number of data against Roc Auc, Prc Ap 
            false_positive_rate, true_positive_rate, thresholds = metrics.roc_curve(test_y[:n_test].reshape((n_test,-1)), pred_y)
            roc_auc = metrics.auc(false_positive_rate, true_positive_rate)

            precision, recall, thresholds = metrics.precision_recall_curve(test_y[:n_test].reshape((n_test,-1)), pred_y)
            average_precision = metrics.average_precision_score(test_y[:n_test].reshape((n_test,-1)), pred_y)
            if plot_ind:
            # Plot individual ROC/AUC
                plt.subplot(1, 2,1)
                plt.title('ROC : AUC = {:.2f}'.format(roc_auc))
                plt.plot(false_positive_rate, true_positive_rate, 'b')
                plt.legend(loc='lower right')
                plt.plot([0,1],[0,1],'r--')
                plt.ylim([-.05, 1.05])
                plt.xlim([-.05, 1.0])
                plt.ylabel('True Positive Rate')
                plt.xlabel('False Positive Rate')
                plt.grid(True)

                # Plot individual PRC/AP
                plt.subplot(1, 2,2)
                plt.plot(recall, precision)
                plt.xlabel('Recall')
                plt.ylabel('Precision')
                plt.ylim([0.0, 1.05])
                plt.xlim([0.0, 1.0])
                plt.title('Precision-Recall : AP={:0.2f}'.format(average_precision))
                plt.legend(loc="lower left")
                plt.grid(True)
                plt.gcf().set_size_inches(10,4)
                if save_individual:
                    plt.savefig("{}/{}_precision_recall_ROC_NoD_{}.pdf".format(save_path, expName, n_data), bbox_inches='tight')
                plt.show()
            
            roc_aucs[i,j] = roc_auc
            prc_aucs[i,j] = average_precision
            
            np.save('{}/{}_roc_aucs.npy'.format(save_path, expName), roc_aucs)
            np.save('{}/{}_prc_aucs.npy'.format(save_path, expName), prc_aucs)
            
            if len(n_datas) == 1 and len(ks) != 1:
                # Plot ROC/AUC & PRC/AP against parameters
                plt.subplot(1, 2, 1)
                plt.grid(True)
                print(i, j,(roc_aucs.shape))
                plt.plot(ks[:j+1], roc_aucs[i,:j+1])
                plt.xlabel('K')
                plt.ylabel('ROC - AUC')
                plt.title('K against ROC, Area Under Curve')
                plt.subplot(1, 2,2)
                plt.grid(True)
                plt.plot(ks[:j+1], prc_aucs[i,:j+1])
                plt.xlabel('K')
                plt.ylabel('PRC - AP')
                plt.title('K against PRC, Average Precision')
                plt.gcf().set_size_inches(10,4)
                plt.savefig("{}/{}_ROC_PRC_AUC.pdf".format(save_path, expName), bbox_inches='tight')
                plt.show()
            elif len(n_datas) != 1 and len(ks) ==1:
                 # Plot ROC/AUC & PRC/AP against parameters
                plt.subplot(1, 2, 1)
                plt.grid(True)
                plt.plot(n_datas[:i+1], roc_aucs[:i+1,j])
                plt.xlabel('Number of Data')
                plt.ylabel('ROC - AUC')
                plt.title('ROC, Area Under Curve')
                plt.subplot(1, 2,2)
                plt.grid(True)
                plt.plot(n_datas[:i+1], prc_aucs[:i+1,j])
                plt.xlabel('Number of Data')
                plt.ylabel('PRC - AP')
                plt.title('PRC, Average Precision')
                plt.gcf().set_size_inches(10,4)
                plt.savefig("{}/{}_ROC_PRC_AUC.pdf".format(save_path, expName), bbox_inches='tight')
                plt.show()
    return roc_aucs, prc_aucs
```

```python
featureset = [3,7,11,12,15,17]
num_files = 1000
threshold = 0.25
train_set_proportion = .75
test_set_proportion = 1 - train_set_proportion
filters = [[11,'>',1000]]
centiles = [0,25,50,75,100]
size = 40
step = 30
reusefeatures = False
savefeatures = True
reuseclassifier = False
saveclassifier = True
saveresults = True

# Split up image files into training and test sets
imgfilenames = glob.glob(IMAGE_DIR + '*.jpg')

baseimgfilenames = [os.path.basename(imgfilenames[i]) 
                    for i in range(num_files)]                            

train, test = train_test_split(np.arange(num_files),
                               train_size=train_set_proportion,
                               test_size=test_set_proportion,
                               random_state=1)  

trainfiles = [baseimgfilenames[i] for i in train]
testfiles = [baseimgfilenames[i] for i in test]  

print 'Extracting features from training images...'
Xtrain_knn = getfeatures(trainfiles, size, step,
                    attributes=featureset,
                    filters=filters,
                    centiles=centiles,
                    loadfromfile=reusefeatures, 
                    savetofile=savefeatures,
                    filename='Xtrain_knn.npy') 

ytrain_knn = getlabels(trainfiles, size, step,
                  loadfromfile=reusefeatures,
                  savetofile=savefeatures,
                  filename='ytrain_knn.npy')     
                            
print 'Extracting features from testing images...'
Xtest_knn_try = getfeatures(testfiles, size, step,
                    attributes=featureset,
                    filters=filters,
                    centiles=centiles,
                    loadfromfile=reusefeatures, 
                    savetofile=savefeatures,
                    filename='Xtest_knn.npy') 

ytest_knn_try = getlabels(testfiles, size, step,
                  loadfromfile=reusefeatures,
                  savetofile=savefeatures,
                  filename='ytest_knn.npy')   

# Classification - Extremely Randomized Trees classfier
if reuseclassifier:
    classifier = pickle.load(open(FEATURES_DIR + 'classifier_knn.pkl', 'rb')) 

else:
    print 'Training...'
    classifier = neighbors.KNeighborsClassifier(n_neighbors=1, weights='distance')
    classifier.fit(Xtrain_knn_try,ytrain_knn_try)

if saveclassifier:
    pickle.dump(classifier, open(FEATURES_DIR + 'classifier_knn.pkl', 'wb'))

print 'Predicting...'
# Plot bounding boxes around parasites annotated on testing images
predictions = classifier.predict_proba(Xtest_knn)[:,1]
for i, img_name in enumerate(testfiles):
    img = (cv2.imread("data/images/{}".format(img_name)))
    output = displayframe(img=img, predictions=predictions[i*850:(i+1)*850], labels=ytest_knn[i*850:(i+1)*850], threshold=threshold, size=size, step=step)
    cv2.imwrite(img_name, output)

    
# Precision-recall curve
precision, recall, thresholds = precision_recall_curve(ytest_knn, predictions)
    
area = auc(recall, precision)
print("Area under precision-recall curve: %0.2f" % area)
    
fig = plt.figure()
fig.set_size_inches(4,4)
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.grid(True)
plt.ylim([0.0, 1.0])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall: AUC=%0.2f' % area)
plt.savefig("precision_recall.pdf")
```

# Extract features from Augmented data

```python
def extract_features_from_patch(patches):
    featureset = [3,7,11,12,15,17]
    filters = [[11,'>',1000]]
    centiles = [0,25,50,75,100]
    feature_ = []
    for i, patch in enumerate(patches):
        if i%1000==0:
            print("{}/{}".format(i,len(patches)))
        feature_.append(shapefeatures.extract(patch[0], featureset, filters, centiles))
        #print(patch[0].shape)
    return feature_
```
